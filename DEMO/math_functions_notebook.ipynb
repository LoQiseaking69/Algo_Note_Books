## Linear Function
The linear function is defined as $f(x) = 2x + 3$. It represents a straight line.

```python

import matplotlib.pyplot as plt
import numpy as np

def linear_function(x):
    return 2*x + 3

x = np.linspace(-10, 10, 100)
plt.plot(x, linear_function(x))
plt.title('Linear Function: $f(x) = 2x + 3$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.savefig('linear.png')
plt.show()
```

## Quadratic Function
The quadratic function is defined as $f(x) = x^2 - 4x + 4$. It represents a parabola.

```python

def quadratic_function(x):
    return x**2 - 4*x + 4

plt.plot(x, quadratic_function(x))
plt.title('Quadratic Function: $f(x) = x^2 - 4x + 4$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.savefig('quadratic.png')
plt.show()
```

## Polynomial Function
The polynomial function is defined as $f(x) = 3x^3 - 2x^2 + x - 5$. It represents a more complex curve.

```python

def polynomial_function(x):
    return 3*x**3 - 2*x**2 + x - 5

plt.plot(x, polynomial_function(x))
plt.title('Polynomial Function: $f(x) = 3x^3 - 2x^2 + x - 5$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.savefig('polynomial.png')
plt.show()
```

## Trigonometric Function
The trigonometric function is defined as $f(x) = sin(x)$. It represents a sine wave.

```python

def trigonometric_function(x):
    return np.sin(x)

x_trig = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.plot(x_trig, trigonometric_function(x_trig))
plt.title('Trigonometric Function: $f(x) = sin(x)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.savefig('trigonometric.png')
plt.show()
```

## Exponential Function
The exponential function is defined as $f(x) = 2^x$. It represents exponential growth.

```python

def exponential_function(x):
    return 2**x

x_exp = np.linspace(-2, 3, 100)
plt.plot(x_exp, exponential_function(x_exp))
plt.title('Exponential Function: $f(x) = 2^x$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.savefig('exponential.png')
plt.show()
```

## Logarithmic Function
The logarithmic function is defined as $f(x) = log(x)$. It represents logarithmic growth.

```python

def logarithmic_function(x):
    return np.log(x)

x_log = np.linspace(0.1, 10, 100)
plt.plot(x_log, logarithmic_function(x_log))
plt.title('Logarithmic Function: $f(x) = log(x)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.savefig('logarithmic.png')
plt.show()
```
