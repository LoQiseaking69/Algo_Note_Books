# Fast Fourier Transform (FFT) Example
This notebook demonstrates the use of the Fast Fourier Transform (FFT) to analyze the frequency components of a signal composed of multiple sine waves.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Signal Generation
Here, we define the parameters for our signal (sampling rate and duration) and generate a signal composed of sine waves of different frequencies.

```python
# Signal parameters
sampling_rate = 1000  # Hz
duration = 1  # second
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# Generate a signal with multiple frequencies
frequencies = [5, 40, 90]  # Hz
signal = sum(np.sin(2 * np.pi * f * t) for f in frequencies)
```

## Applying FFT
The Fast Fourier Transform (FFT) is applied to the signal to convert it from the time domain to the frequency domain.

```python
# Apply FFT
fft_result = np.fft.fft(signal)
freqs = np.fft.fftfreq(len(signal), d=1/sampling_rate)
```

## Time Domain Visualization
Let's plot the time-domain representation of our signal.

```python
# Plot the signal
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t, signal)
plt.title('Time Domain Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
```

## Frequency Domain Visualization
Now, we plot the magnitude spectrum to visualize the frequency components of our signal.

```python
# Plot the magnitude spectrum
plt.subplot(2, 1, 2)
plt.plot(freqs, np.abs(fft_result))
plt.title('Magnitude Spectrum')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.xlim(0, 150)  # Limit x-axis for better visibility

plt.tight_layout()
plt.show()
```

# Conclusion
In this notebook, we generated a multi-frequency signal, applied FFT to analyze its frequency components, and visualized the results. This demonstrates how FFT can be used to extract frequency information from a time-domain signal.
